```python
import pandas as pd

# replace with your filename
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/data/CaseKey.csv")
df.head()
```

```python
df["Interest_Business"] = df["Հետաքրքրվա՞ծ եք բիզնես դասընթացներով/ծրագրերով։"].map({"այո": 1, "ոչ": 0})
```

```python
df["WTP"] = df["Պատրաստ կլինեի՞ք վճարել  նմանատիպ դասընթացի համար, եթե այն համապատասխաներ Ձեր ակնկալիքներին:"].map({"այո": 1, "ոչ": 0})
```

```python
import numpy as np
def clean_price(x):
    if pd.isna(x):
        return np.nan
    x = str(x).replace("դրամ", "").replace(",", "").replace(" ", "")
    digits = "".join([c for c in x if c.isdigit()])
    return int(digits) if digits else np.nan

df["WTP_Amount"] = df["Նշեք գինը, որ պատրաստ եք վճարել (ամսական,դրամ)"].apply(clean_price)
```

```python
factors_col = "Ո՞ր գործոնները առավելապես կարող էին ազդել Ձեր՝ դասընթացին  մասնակցելու որոշման վրա։ (առավելագույնը 4 տարբերակ)"

factor_options = [
    "Դասընթացի գինը",
    "Աշխատանքի հնարավորությունները",
    "Վկայականի առկայությունը",
    "Ցանցային կապերի հնարավորություն (networking)",
    "Ծրագրի բովանդակությունը / թեմաները",
    "Դասավանդողների փորձն ու հեղինակությունը",
    "Ընկերների / ծանոթների խորհուրդը",
    "Դասընթացի տևողությունը"
]

for f in factor_options:
    df["Factor_" + f] = df[factors_col].apply(lambda x: 1 if pd.notna(x) and f in str(x) else 0)
```

```python
likelihood_map = {
    "Հիմնականում հավանական չէ": 1,
    "Մասամբ հավանական չէ": 2,
    "Մասամբ հավանական է": 3,
    "Հիմնականում հավանական է": 4
}
df["Likelihood"] = df["Որքանո՞վ է հավանական, որ հենց հիմա կկատարեիք վճարում:"].map(likelihood_map)
```

```python
Q1 = df["WTP_Amount"].quantile(0.25)
Q3 = df["WTP_Amount"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df["WTP_Amount_Clean"] = df["WTP_Amount"].where(
    (df["WTP_Amount"] >= lower_bound) & (df["WTP_Amount"] <= upper_bound)
)
```

```python
# Preview cleaned dataset
print(df[["Interest_Business", "WTP", "WTP_Amount", "WTP_Amount_Clean", "Likelihood"] + [c for c in df.columns if c.startswith("Factor_")]].tail())
```

```python
# Filter for people interested in business courses.
interested_df = df[df["Interest_Business"] == 1]

# Percentage of interested people willing to pay.
willing_to_pay_percentage = (interested_df["WTP"].sum() / len(interested_df)) * 100

# Average WTP amount for those interested.
average_wtp_interested = interested_df["WTP_Amount_Clean"].mean()

# Analyze the factors valued by interested people.
factor_analysis = interested_df[[c for c in interested_df.columns if c.startswith("Factor_")]].sum().sort_values(ascending=False)
factor_analysis_percentage = (factor_analysis / len(interested_df)) * 100

# Map selection status
selection_status_map = {
    "Այո": "Passed",
    "Ոչ": "Not Passed",
    "Մասամբ": "Partially Passed"
}
df["Selection_Status"] = df["Անցե՞լ եք CaseKey-ի ընտրության փուլը:"].map(selection_status_map)

# Filter for people who did not pass or partially passed the selection phase.
non_pass_df = df[(df["Selection_Status"] == "Not Passed") | (df["Selection_Status"] == "Partially Passed")]

# Analyze willingness to pay and amount for this group.
non_pass_wtp_percentage = (non_pass_df["WTP"].sum() / len(non_pass_df)) * 100
non_pass_wtp_amount = non_pass_df["WTP_Amount_Clean"].mean()

# Analyze likelihood to pay for this group.
non_pass_likelihood_df = non_pass_df.groupby("Likelihood")["Likelihood"].count()
non_pass_likelihood_distribution = (non_pass_likelihood_df / non_pass_likelihood_df.sum()) * 100

# Print the results.
print(f"Percentage of people interested in business courses who are willing to pay: {willing_to_pay_percentage:.2f}%")
print(f"Average amount people interested in business courses are willing to pay: {average_wtp_interested:.2f} AMD")
print("\nTop 3 Factors for people interested in business courses:")
print(factor_analysis_percentage.head(3))
print("\nAnalysis for people who did not pass or partially passed the selection phase:")
print(f"Percentage of this group willing to pay for a course: {non_pass_wtp_percentage:.2f}%")
print(f"Average amount this group is willing to pay: {non_pass_wtp_amount:.2f} AMD")
print("\nLikelihood to pay distribution for this group:")
print(non_pass_likelihood_distribution)
```
