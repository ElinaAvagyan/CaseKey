```python

import pandas as pd

# replace with your filename
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/data/yandex.csv")
df.head()
```

```python
def find_col(cols, startswith_text):
    for c in cols:
        if c.strip().startswith(startswith_text):
            return c
    raise ValueError(f"Column starting with '{startswith_text}' not found.")

# columns
freq_col = find_col(df.columns, "Հաճա՞խ եք այցելում ժամանցային միջոցառումների")
money_col = find_col(df.columns, "Որքա՞ն եք ամսական վճարում ժամանցի համար")

# frequencies to include
valid_freq = ["Ամեն օր", "Ամեն շաբաթ", "Ամսը մեկ անգամ"]
sub = df[df[freq_col].isin(valid_freq)].copy()

# ---- EXCLUDE these ranges manually ----
exclude_ranges = [
    "30,100 – 60,000 դր.", "60,100 – 100,000 դր.",
    "30.100 – 60.000 դր.", "60.100 – 100.000 դր."
]
sub = sub[~sub[money_col].isin(exclude_ranges)]

# ---- parse ranges into numeric ---
def to_int_thousands(s):
    return int(s.replace(".", "").replace(",", ""))

def parse_money(text):
    text = str(text)
    nums = [n for n in re.findall(r"[\d\.,]+", text) if re.sub(r"[^\d]", "", n)]
    vals = [to_int_thousands(n) for n in nums]

    if len(vals) >= 2:
        return (vals[0] + vals[1]) / 2
    elif len(vals) == 1:
        return vals[0] / 2
    return None

sub["money"] = sub[money_col].apply(parse_money)
sub = sub.dropna(subset=["money"])

# --- IQR outlier removal ---
q1 = sub["money"].quantile(0.25)
q3 = sub["money"].quantile(0.75)
iqr = q3 - q1

lo = q1 - 1.5 * iqr
hi = q3 + 1.5 * iqr

clean = sub[(sub["money"] >= lo) & (sub["money"] <= hi)]

mean_spent = clean["money"].mean()

print(f"Mean monthly spending (excluding 30,100–60,000 & 60,100–100,000): {round(mean_spent, 2)} դրամ")
print(f"Included responses: {len(clean)} from {len(sub)}")
```
